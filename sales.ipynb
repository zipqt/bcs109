# Sales

This worksheet will calculate totals based on different shopping conditions.


1.   The first section takes the user's quantity, determines the discount rewarded, and calculates and displays the total after the discount is applied.
2.   The second section takes the user's items weight, and displays the cost to ship the item.




*   quantity
*   item weight




```python
package = 99
q = int(input('How many would you like to purchase?'))
discount = 0
if q >= 10 and q <= 19:
  discount = .1
elif q >=20 and q <= 49:
  discount = .2
elif q >=50 and q <= 99:
  discount = .3
elif q>=100:
  discount = .4

total = package*q*(1-discount)
print('The discount saved is ', discount*100, "%")
print('The total amount of the current cost is $',total)
```

```python
w = float(input('How heavy is the package you are shipping? '))
rate = 1.5
if w <= 2:
  rate = 1.5
elif w > 2 and w <= 6:
  rate = 3.0
elif w > 6 and w <= 10:
  rate = 4.0
elif w > 10:
  rate = 4.75

t = w*rate
total = "{:.2f}".format(t)

print('The cost to ship your package will be $', total)
```
