First section asks user for age and name, only accepts string for name and int for age.

Second section prints all numbers from 1 to 100 that include a 3.

Third section counts all even numbers backwards from 100.

Fourth section prints all prime numbers between 100 and 100.

```python
name = '1234'
age = -1
check = 0

while(True):
  check+=1
  if check > 3:
    print('unacceptable')
    break

  name = input('please enter your name: ')
  age = input('please enter your age: ')

  if(age.isdigit() and all(x.isalpha() or x.isspace() for x in name) and (int(age) >= 0 and int(age) <= 100)):
    print('Acceptable')
    break
```

```python
for x in range(1, 100):
  val = str(x)
  for position in val:
    if position == '3':
      print(x)
      break
```

```python
for number in range(100,0,-1):
  if number % 2 == 0:
    print(number)
```

```python
start = 100
end = 1000
  
for i in range(start, end+1):
  if i>1:
    for j in range(2,i):
        if(i % j==0):
            break
    else:
        print(i,end="|")
```
