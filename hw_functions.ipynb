```python
#takes two numbers and prints the summ of all the mumberes between to the two values.

def func2(x,y):
    sum = x
    for i in range(x,y):
        sum = sum + i+1
    return sum

num1 = int(input("Give me a number: "))
num2 = int(input("Give me a number: ")) 
print(func2(num1,num2))

```

```python
#takes an item's cost, discount, and tax rate and outputs the total cost after tax and discount is applied.

def func3():
  item_cost = float(input("Enter the item's original cost: "))
  percentage_off = float(input("Enter the item's discount in percent form: "))/100
  tax = float(input("Enter the Tax rate in percent form: "))/100

  discount = item_cost * percentage_off
  discounted_price = item_cost - discount
  taxes = discounted_price * tax
  total = discounted_price + taxes
  formated_total = "{:.2f}".format(total)
  print("The total will be: $",formated_total)

func3()


```

```python
#takes two numbers and prints all the primes between the two numbers.

def isPrime(num):
  for i in range(2,num):
    if num % i == 0:
      return False
    return True

def listPrimes(x,y):
  for i in range(x,y+1):
    if (isPrime(i)):
      print(i, end = " | ")

num1 = int(input('Give me a number: '))
num2 = int(input('Give me a number: '))

listPrimes(num1,num2)
```

```python
#takes a number and lists the coprimes from one up until the givin number.

def coPrime_test(num1,num2):
  for i in range(2,num2+1):
    if num1%i == 0 and num2%i == 0:
      return False
  return True

def list_coprimes(x):
  print(1)
  for i in range(2,x):
    if coPrime_test(x,i):
      print(i)

num = int(input("Give me a number:"))
list_coprimes(num)
```

```python
#takes two numbers and lists the twin primes between the two numbers.

def isTwinPrime(x):
  if isPrime(x) and isPrime(x+2):
    return True
  return False

def listTwinPrimes(x,y):
  for i in range(x,y+1):
    if isTwinPrime(i):
      print(i, " ", i+2)
      i+=2

num1 = int(input("Give me a number: "))
num2 = int(input("Give me a number: "))
listTwinPrimes(num1,num2)
```

```python
#plays rock paper scissors 10000 times and prints the odds of rolling each option.

import random as r

def rps():
  rock_count = 0
  paper_count = 0
  scissors_count = 0

  for i in range(10000):
    num1 = r.randint(1,3)
    num2 = r.randint(1,3)
    if num1 == num2:
      continue
    elif num1 == 1 and num2 == 2:
      paper_count +=1
    elif num1 == 1 and num2 == 3:
      rock_count +=1
    elif num1 == 2 and num2 == 3:
      scissors_count+=1
  print("rock:", rock_count/10000)
  print("paper:", paper_count/10000)
  print("scissors:", scissors_count/10000)

rps()
```
