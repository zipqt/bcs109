https://archive.ics.uci.edu/ml/datasets/Blood+Transfusion+Service+Center

Data from 749 randomly selected donors from Blood Transfusion Service Center in Hsin-Chu City in Taiwan. Each donor has 5 attributes-
R (Recency - months since last donation),
F (Frequency - total number of donation),
M (Monetary - total blood donated in c.c.),
T (Time - months since first donation), 
a binary variable representing whether he/she donated blood in March 2007 (1 stand for donating blood; 0 stands for not donating blood).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#r = months since last donation
#f = total number of donation
#cc = total blood donated in c.c
#t = months since first donation
#y/n 1 if donated blood, 2 if not

df = pd.read_csv("https://raw.githubusercontent.com/zipqt/bcs109/main/transfusion.data")
df2 = df.rename({'Recency (months)': 'R', 'Frequency (times)': 'F','Monetary (c.c. blood)':'cc', 'Time (months)':'T','whether he/she donated blood in March 2007':'y/n'}, axis=1)
df2.head(10)
```

```python
#check if we have any empty data
df2.isna().sum()
```

```python
#check how many unique values we have in each column
df2.nunique()
```

```python
#shows how many people donated a certain amount of blood
#first column represents blood donated in ccs
#second column represents how many people donated that amount of blood
df2['cc'].value_counts()
#df2['cc'].value_counts().plot(kind='bar', fontsize = 12)
```

```python
#0 represents first time donors 
#1 represents people who have donated blood before

#clearly shows the majority of people did not donante blood prior to the month the data was taken from
df2['y/n'].value_counts().plot(kind='bar', fontsize = 12)
```

After analyzing the data, two discoveries I made were that the most common amount of blood donated ranged from 250 to 1000 ccs, and that it was much more common for people to be first time donors.
